```python
import os

import cv2
#from google.colab.patches import cv2_imshow

import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

# For drawing landmarks
from mediapipe import solutions
from mediapipe.framework.formats import landmark_pb2
import numpy as np
```

```python

#Function for drawing landmarks
def draw_landmarks_on_image(rgb_image, detection_result):
  pose_landmarks_list = detection_result.pose_landmarks
  annotated_image = np.copy(rgb_image)

  # Loop through the detected poses to visualize.
  for idx in range(len(pose_landmarks_list)):
    pose_landmarks = pose_landmarks_list[idx]

    # Draw the pose landmarks.
    pose_landmarks_proto = landmark_pb2.NormalizedLandmarkList()
    pose_landmarks_proto.landmark.extend([
      landmark_pb2.NormalizedLandmark(x=landmark.x, y=landmark.y, z=landmark.z) for landmark in pose_landmarks
    ])
    solutions.drawing_utils.draw_landmarks(
      annotated_image,
      pose_landmarks_proto,
      solutions.pose.POSE_CONNECTIONS,
      solutions.drawing_styles.get_default_pose_landmarks_style())
  return annotated_image

def findVector2(image):
    '''
    Takes in an image and outputs the detection results
    '''
    model_path = 'models\\pose_landmarker_lite.task'

    '''
        image: path to image OR numpyp image
        -> vectorization of image
    '''
    BaseOptions = mp.tasks.BaseOptions
    PoseLandmarker = mp.tasks.vision.PoseLandmarker
    PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
    VisionRunningMode = mp.tasks.vision.RunningMode

    options = PoseLandmarkerOptions(
        base_options=BaseOptions(model_asset_path=model_path),
        running_mode=VisionRunningMode.IMAGE)

    with PoseLandmarker.create_from_options(options) as landmarker:

        #Image input from file
        mp_image = mp.Image.create_from_file(image)
        #Image input from numpy array (for movies)
        #mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=numpy_image)


        detection_result = landmarker.detect(mp_image)

        #Uncomment to display resultant image
        #original_np_image = mp_image.numpy_view()
        #original_np_image = cv2.resize(original_np_image, (640, 480))  # (width, height)

        # Draw landmarks
        #annotated_image = draw_landmarks_on_image(original_np_image, detection_result)

        # Show using OpenCV
        #cv2.imshow('Pose', cv2.cvtColor(annotated_image, cv2.COLOR_RGB2BGR))
        #cv2.waitKey(0)
        cv2.destroyAllWindows()


    return detection_result

```

```python
import json
def save_pose_landmarks_to_json(pose_landmarker_result, output_filename='pose_landmarks.json'):
    # Create a dictionary to store landmarks and other information
    pose_data = {
        "landmarks": []
    }
    
    # Extract landmarks from PoseLandmarkerResult
    for landmark in pose_landmarker_result.landmarks:
        landmark_data = {
            "x": landmark.x,
            "y": landmark.y,
            "z": landmark.z,
            "visibility": landmark.visibility
        }
        pose_data["landmarks"].append(landmark_data)
    
    # Convert the Python dictionary to a JSON string and save it to a file
    with open(output_filename, 'w') as json_file:
        json.dump(pose_data, json_file, indent=4)
```
